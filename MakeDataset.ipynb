```python
import pandas as pd
import os
from datasets import Dataset, Audio

# Path to your CSV file and audio files
csv_path = '/content/drive/My Drive/dataset.csv'  # update the path
audio_folder = '/content/drive/My Drive/AudioFiles/'  # update the path

# Load the CSV file
df = pd.read_csv(csv_path)

# Verify the contents of the CSV file
print(df.head())
```

```python
# Create a dictionary to map serial numbers to audio file paths
audio_files = {i: os.path.join(audio_folder, f'audio{i}.mp3') for i in range(1, 51)}

# Add the audio file paths to the DataFrame
df['audio'] = df['Seial number'].map(audio_files)

# Drop the 'Serial number' column
df = df.drop(columns=['Seial number'])

# Verify the contents of the updated DataFrame
print(df.head())
```

```python
# Convert the DataFrame to a Hugging Face Dataset
dataset = Dataset.from_pandas(df)

# Cast the 'audio' column to Audio
dataset = dataset.cast_column('audio', Audio())

# Verify the dataset
print(dataset)

# Upload the dataset to Hugging Face
from huggingface_hub import login

# Log in to Hugging Face
login()

```

```python

# Upload the dataset
dataset.push_to_hub("PradyumSomebody/Audio_Medical5")
```
